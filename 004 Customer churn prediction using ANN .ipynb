### **6.1 Method 1 : Undersampling**

```python
# Class count 
count_class_0,count_class_1 = df1.Churn.value_counts()

# Divide by class
df_class_0 = df2[df2['Churn']==0]
df_class_1 = df2[df2['Churn']==1]
```

```python
print(count_class_0,count_class_1)
```

```python
df_class_0.shape
```

```python
df_class_1.shape
```

```python
df_class_0_under = df_class_0.sample(count_class_1)
```

```python
df_class_0_under.shape
```

```python
# Now it is important to concat both the above dataframes
df_test_under = pd.concat([df_class_0_under,df_class_1], axis = 0)
```

```python
df_test_under.shape
```

```python
df_test_under['Churn'].value_counts()
```

### **Create X and Y from this new Dataframe**

```python
X = df_test_under.drop('Churn', axis='columns')
Y = df_test_under['Churn']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size =0.2, random_state=2, stratify=Y)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
print(Y.shape, Y_train.shape, Y_test.shape)
```

```python
Y_train.value_counts()
```

```python
Y_test.value_counts()
```

```python
# Defining ANN Model
def ANN(X_train, Y_train, X_test,Y_test, loss, weights):
  model = keras.Sequential([
    keras.layers.Dense(26, input_shape=(26,), activation = 'relu'),
    keras.layers.Dense(15,  activation = 'relu'),
    keras.layers.Dense(1,  activation = 'sigmoid'),
    ])
  

  model.compile(optimizer = 'adam',loss = loss,metrics = ['accuracy'])

  if weights == -1:
    model.fit(X_train, Y_train, epochs = 100)
  else:
    model.fit(X_train, Y_train, epochs = 100, class_weights = weights)

  print(model.evaluate(X_test, Y_test))

  Y_preds = model.predict(X_test)
  Y_preds = np.round(Y_preds)

  print("Classification report: \n", classification_report(Y_test, Y_preds))

  return Y_preds
```

```python
Y_preds = ANN(X_train, Y_train, X_test,Y_test, 'binary_crossentropy', -1)
```

```python
df3=X_train.copy()
df3['Churn']=Y_train
```

```python
df3_class_0 = df3[df3['Churn']==0]
df3_class_1 = df3[df3['Churn']==1]
```

```python
df3_class_0.shape, df3_class_1.shape
```

```python
def get_trained_batch(df_majority,df_minority,start,end):
  df_train = pd.concat([df_majority[start:end],df_minority], axis=0)     #major class

  X_train = df_train.drop('Churn', axis='columns')
  Y_train = df_train['Churn']

  return X_train, Y_train
```

```python
X_train, Y_train = get_trained_batch(df3_class_0, df3_class_1, 0,1495)

Y_pred1 = ANN(X_train, Y_train, X_test, Y_test, 'binary_crossentropy',-1)
```

```python
X_train, Y_train = get_trained_batch(df3_class_0, df3_class_1,1495, 2990)

Y_pred2 = ANN(X_train, Y_train, X_test, Y_test, 'binary_crossentropy',-1)
```

```python
X_train, Y_train = get_trained_batch(df3_class_0, df3_class_1,2990, 4130)

Y_pred3 = ANN(X_train, Y_train, X_test, Y_test, 'binary_crossentropy',-1)
```
